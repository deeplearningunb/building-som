# Self Organizing Maps

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
# http://archive.ics.uci.edu/ml/datasets/statlog+(australian+credit+approval)
dataset = pd.read_csv('Credit_Card_Applications.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
dataset.head(15)
# dataset.iloc[:, 1:-1]
```

```python
# Feature Scaling
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0, 1))
X = sc.fit_transform(X)
X
```

```python
# Training the SOM
from minisom import MiniSom
som = MiniSom(x = 10, y = 10, input_len = 15, sigma = 1.0, learning_rate = 0.5)
som.random_weights_init(X)
som.train_random(data = X, num_iteration = 100)
```

```python
# Visualizing the results
from pylab import bone, pcolor, colorbar, plot, show
bone()
pcolor(som.distance_map().T)
colorbar()
show()
```

```python
# Visualizing the results
from pylab import bone, pcolor, colorbar, plot, show
bone()
pcolor(som.distance_map().T)
colorbar()
markers = ['o', 's']
colors  = ['r', 'g']
for i, x in enumerate(X):
    w = som.winner(x)
    plot(w[0] + 0.5,
         w[1] + 0.5,
         markers[y[i]],
         markeredgecolor = colors[y[i]],
         markerfacecolor = 'None',
         markersize = 10,
         markeredgewidth = 2)
show()
```

```python
# Finding the frauds
mappings = som.win_map(X)
frauds = np.concatenate((mappings[(0,8)], mappings[(2,9)]), axis = 0)

frauds = sc.inverse_transform(frauds)
for client in frauds:
    print(client[0])
```

# SimpSOM

```python
# ! pip install SimpSOM
import pandas as pd
import SimpSOM as sps
from sklearn.cluster import KMeans
import numpy as np

net = sps.somNet(20, 20, X, PBC=True)
net.train(0.5, 100)
# net.save(filename_weights)
net.nodes_graph(colnum=0)
```

```python
net.diff_graph()
```

```python
#Project the datapoints on the new 2D network map.
projection = net.project(X)
```

```python
projection = np.array(projection)
```

```python
list_int = lambda x:x[0]

projX = [list_int(x) for x in projection[:,:1]]
projY = [list_int(x) for x in projection[:,1:]]
```

```python
plt.scatter(projX,projY)
plt.show()
```

```python
#Cluster the datapoints according to the Quality Threshold algorithm.
cluster = net.cluster(X, type='qthresh',show=True)
```
